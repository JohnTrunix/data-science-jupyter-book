Also called: Angular Range

A range area chart is a variation of an area chart that lets you plot bands of data, such as Bollinger bands and weather patterns. Each point in the chart is specified by two y values.

`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
`````

`````python
data = {'X': np.arange(12),
        'Value1': np.array([5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4]),
        'Value2': np.array([6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8])}
df = pd.DataFrame(data, columns=['X', 'Value1', 'Value2'])
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#023E7D', '#FBD87F', '#829CBC']
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(df.X, df.Value1, color=colors[0], label='Value1')
ax.plot(df.X, df.Value2, color=colors[1], label='Value2')
ax.fill_between(df.X, df.Value1, df.Value2, where=(df.Value1 > df.Value2), interpolate=True, color=colors[2], alpha=0.5)
ax.fill_between(df.X, df.Value1, df.Value2, where=(df.Value1 < df.Value2), interpolate=True, color=colors[3], alpha=0.5)
plt.show()
`````
