Also called: Column Sparkline Chart

A Sparkline is a small intense, simple, word-sized graphic with typographic resolution. Sparklines mean that graphics are no longer cartoonish special occasions with captions and boxes, but rather sparkline graphics can be everywhere a word or number can be: embedded in a sentence, table, headline, map, spreadsheet, graphic. Data graphics should have the resolution of typography. A Column Sparkline is basically same as the classic sparkline, except that it uses columns/bars instead of lines.

`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
`````

`````python
np.random.seed(14)
df = pd.DataFrame(np.random.randint(0,100, size=(100, 5)), columns=['A','B','C','D','E'])
df.head()
`````

`````python
colors = ['#002855', '#FBD87F', '#023E7D', '#376996', '#829CBC', '#F8C23A']
fig, ax = plt.subplots(len(df.columns), 1, figsize=(10, 6))
fig.patch.set_facecolor('white')

for i, col in enumerate(df.columns):
    ax[i].plot(df[col], color=colors[i])
    for k,v in ax[i].spines.items():
        v.set_visible(False)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].plot(len(df[col]) - 1, df[col][len(df[col]) - 1], 'r.')
    ax[i].fill_between(range(len(df[col])), df[col], len(df[col])*[min(df[col])], color=colors[i], alpha=0.4)
# set for last plot xticks
ax[-1].set_xticks(range(0, len(df[col])+1, 10))
plt.show()
`````
