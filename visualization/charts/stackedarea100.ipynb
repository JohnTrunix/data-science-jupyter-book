`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
`````

`````python
df = pd.DataFrame(np.random.random((30,6)), columns=['Val1', 'Val2', 'Val3', 'Val4', 'Val5', 'Val6'] ,index=np.arange(1, 31))
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#023E7D', '#376996', '#829CBC', '#FBD87F', '#F8C23A']
fig, ax = plt.subplots(figsize=(8, 6))

data_percentage = df.divide(df.sum(axis=1), axis=0)

ax.stackplot(df.index, data_percentage.Val1, data_percentage.Val2, data_percentage.Val3, data_percentage.Val4, data_percentage.Val5, data_percentage.Val6, colors=colors)
plt.xlim(df.index.min(), df.index.max())
plt.ylim(0, 1)
plt.show()
`````
