Bubble charts can facilitate the understanding of social, economical, medical, and other scientific relationships. Bubble charts can be considered a variation of the scatter plot, in which the data points are replaced with bubbles.

`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
`````

`````python
df = sns.load_dataset('iris')
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#023E7D', '#376996', '#829CBC', '#FBD87F', '#F8C23A']
fig = plt.figure(figsize=(8, 6))

sns.scatterplot(data=df, x='sepal_length', y='sepal_width', size='petal_width', sizes=(1, 400), 
    hue='petal_length', palette=mpl.colors.ListedColormap(colors), alpha=0.6)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', fontsize=16)
plt.show()
`````
