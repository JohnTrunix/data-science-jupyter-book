`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
`````

`````python
df = pd.DataFrame(np.random.random((30,6)), columns=['Val1', 'Val2', 'Val3', 'Val4', 'Val5', 'Val6'] ,index=np.arange(1990, 2020))
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#023E7D', '#376996', '#829CBC', '#FBD87F', '#F8C23A']
fig, ax = plt.subplots(figsize=(8, 6))

ax.stackplot(df.index, df.Val1, df.Val2, df.Val3, df.Val4, df.Val5, df.Val6, baseline="sym", colors=colors)
plt.xlim(df.index.min(), df.index.max())
plt.show()
`````
