A radar chart is a graphical method of displaying multivariate data in the form of a two-dimensional chart of three or more quantitative variables represented on axes starting from the same point. The relative position and angle of the axes is typically uninformative.

`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from math import pi
`````

`````python
np.random.seed(14)
df = pd.DataFrame(np.random.randint(1,10, size=(3,5)), columns=['Val1', 'Val2', 'Val3', 'Val4', 'Val5'], index=['A', 'B', 'C']).reset_index()
df = df.rename(columns={'index': 'Group'})
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#FBD87F', '#376996']
fig = plt.figure(figsize=(8, 8))

categories=list(df)[1:]
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1], categories)
ax.set_rlabel_position(0)
plt.yticks(color='black')


for i in range(len(df)):
    values=df.loc[i].drop('Group').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid', color=colors[i], label=df.Group[i])
    ax.fill(angles, values, colors[i], alpha=0.2) 

plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
`````
