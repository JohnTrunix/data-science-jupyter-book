Also called: Circular Bar Chart

A Radial/Circular Bar Chart simply refers to a typical Bar Chart displayed on a polar coordinate system, instead of a cartesian system. It is used to show comparisons among categories by using a circular shape.

`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import math
`````

`````python
data = {'Group': ['A', 'B', 'C', 'D'],
        'Value': [150, 300, 220, 100]}
df = pd.DataFrame(data, columns=['Group', 'Value'])
df.head()
`````

`````python
mpl.style.use('seaborn')
colors = ['#002855', '#FBD87F', '#023E7D', '#376996', '#829CBC', '#F8C23A']

plt.figure(figsize=(6, 6))
ax = plt.subplot(projection='polar')
for i in range(len(df)):
    ax.barh(y=i, width=math.radians(df.iloc[i]['Value']), color=colors[i])
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_rlabel_position(0)
ax.set_thetagrids([0, 90, 180, 270], labels=[0, 90, 180, 270])
ax.set_rgrids([0, 1, 2, 3], labels=df['Group'])
plt.show()
`````
