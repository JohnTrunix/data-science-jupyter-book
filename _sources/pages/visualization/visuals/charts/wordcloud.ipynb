`````python
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
`````

`````python
data ={'Name': ['Python', 'Python', 'Python', 'Matplotlib', 'Matplotlib', 'Seaborn', 'Network', 'Plot', 'Violin', 
        'Chart', 'Pandas', 'Datascience', 'Wordcloud', 'Spider', 'Radar', 'Parrallel', 'Alpha', 'Color', 'Brewer', 
        'Density', 'Scatter', 'Barplot', 'Barplot', 'Boxplot', 'Violinplot', 'Treemap', 'Stacked', 'Area', 'Chart', 
        'Chart', 'Visualization', 'Dataviz', 'Donut', 'Pie', 'Time-Series', 'Wordcloud', 'Wordcloud', 'Sankey', 'Bubble']}
df = pd.DataFrame(data, columns=['Name'])
df.head()
`````

`````python
colors = ['#F8C23A', '#FBD87F', '#002855', '#023E7D', '#376996', '#829CBC']

wordcloud = WordCloud(width=800, height=400, background_color='white', colormap=mpl.colors.ListedColormap(colors)).generate(str(df['Name']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
`````
